```python
data['dateRegistered'] = data['yearOfRegistration'] + data['monthOfRegistration']/12
```

```python
def calculateDate(str):
    y = int(str[:4])
    m = int(str[5:7])
    return y+m/12
data['dateAdvertised'] = data['dateCreated'].apply(calculateDate)
```

```python
data['age'] = data['dateAdvertised'] - data['dateRegistered']
data.head()
```

```python
del data['dateRegistered']
del data['dateAdvertised']
del data['yearOfRegistration']
del data['monthOfRegistration']
del data['nrOfPictures']
del data['dateCreated']
```

```python
set(data['seller'])
```

```python
def getSeller(str):
    if str.startswith('p'):
        return 1
    else:
        return 2
data['seller'] = data['seller'].apply(getSeller)
```

```python
set(data['offerType'])
```

```python
def getOffer(str):
    if str.startswith('A'):
        return 1
    else:
        return 2
data['offerType'] = data['offerType'].apply(getOffer)
```

```python
data.head()
```

```python
set(data['abtest'])
```

```python
def getAbtest(str):
    if str.startswith('c'):
        return 1
    else:
        return 2
data['abtest'] = data['abtest'].apply(getAbtest)
```

```python
data.describe(include='all')
```

```python
data['vehicleType'] = data['vehicleType'].fillna('limousine')
```

```python
data.describe(include='all')
```

```python
data['gearbox'] = data['gearbox'].fillna('manuell')
```

```python
data['fuelType'] = data['fuelType'].fillna('benzin')
```

```python
data['notRepairedDamage'] = data['notRepairedDamage'].fillna('nein')
```

```python
data.dropna(inplace=True)
```

```python
data.head()
```

```python
set(data['vehicleType'])
```

```python
def getVehicleType(str):
    if str == 'andere':
        return 1
    elif str == 'bus':
        return 2
    elif str == 'cabrio':
        return 3
    elif str == 'coupe':
        return 4
    elif str == 'kleinwagen':
        return 5
    elif str == 'kombi':
        return 6
    elif str == 'limousine':
        return 7
    else:
        return 8
data['vehicleType'] = data['vehicleType'].apply(getVehicleType)
```

```python
data.head()
```

```python
def getGearbox(str):
    if str.startswith('m'):
        return 1
    else:
        return 2
data['gearbox'] = data['gearbox'].apply(getGearbox)
```

```python
set(data['fuelType'])
```

```python
def getFuelType(str):
    if str == 'andere':
        return 1
    elif str == 'benzin':
        return 2
    elif str == 'cng':
        return 3
    elif str == 'diesel':
        return 4
    elif str == 'elektro':
        return 5
    elif str == 'hybrid':
        return 6
    else:
        return 7
data['fuelType'] = data['fuelType'].apply(getFuelType)
```

```python
def getnotRepairedDamage(str):
    if str == 'nein':
        return 1
    else:
        return 2
data['notRepairedDamage'] = data['notRepairedDamage'].apply(getnotRepairedDamage)
```

```python
models = set(data['model'])
modelNos = {}
i = 1
for model in models:
    modelNos[model] = i
    i += 1
```

```python
def getModel(str):
    return modelNos[str]
data['model'] = data['model'].apply(getModel)
```

```python
brands = set(data['brand'])
brandNo = {}
i = 1
for brand in brands:
    brandNo[brand] = i
    i += 1
```

```python
def getBrand(str):
    return brandNo[str]
data['brand'] = data['brand'].apply(getBrand)
```

```python
data.head()
```

```python
data.head()
```

```python
for vt in set(data['vehicleType']):
    data['vehicleType'+str(vt)] = (data['vehicleType'] == vt).astype(int)
```

```python
data.head()
```

```python
del data['vehicleType']
```

```python
data['brandModel'] = data['brand'].astype(str) + data['model'].astype(str)
```

```python
del data['brand']
del data['model']
```

```python
for bm in set(data['brandModel']):
    data['brandModel'+bm] = (data['brandModel'] == bm).astype(int)
```

```python
del data['brandModel']
```

```python
for ft in set(data['fuelType']):
    data['fuelType'+str(ft)] = (data['fuelType'] == ft).astype(int)
```

```python
del data['fuelType']
```

```python
price = data['price']
```

```python
m = price.median()
price = price.replace(0,m)
```

```python
data['finalPrice'] = price
```

```python
del data['price']
```

```python
del data['seller']
del data['offerType']
del data['abtest']
```

```python
data = data[(data['age'] > 0) & (data['age'] < 70)]
```

```python
data = data[(data.finalPrice > 100) & (data.finalPrice < 150000)]
```

```python
data.values.shape
```

```python
x = data.values[:,:-1]
y = data.values[:,-1]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0)
```

```python
sc = MinMaxScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
sgd = SGDClassifier(loss="squared_hinge", penalty='l2',max_iter=1e-3)
```

```python
sgd.fit(x_train,y_train)
sgd.score(x_test, y_test)
```
